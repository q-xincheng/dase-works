1.编写 Python程序,判断输入a是否为质数

```python
a = int(input("输入整数a: "))
if a <= 1: 
    print("a不是质数")
elif a == 2: 
    print("a是质数")
else: 
    for i in range(2, a):
        if a % i == 0: 
            print("a不是质数")
            break
    else: 
        print("a是质数")
```

2.用Python实现插入排序该算法，不允许直接调用插入排序函数，需要手动实现其逻辑过程。

```python
L_str = input("请输入一组整数，用空格分隔: ")
L = list(map(int, L_str.split(' ')))
for i in range(1,len(L)):
    temp = L[i]
    j = i - 1
    while j >= 0 and L[j] > temp:
        L[j+1] = L[j]
        j -= 1
    L[j+1] = temp
print(L)
```

3.用Python实现选择排序该算法，不允许直接调用插入排序函数，需要手动实现其逻辑过程。

```python
def selction_sort(L):
    for i in range(len(L)-1):
        min_index = i
        for j in range(i+1, len(L)):
            if L[j] < L[min_index]:
                min_index = j
        L[i], L[min_index] = L[min_index], L[i]
    return L

L_str = input("请输入一组整数，用空格分隔: ")
L = list(map(int, L_str.split(' ')))
print(selction_sort(L))
```

4.用Python实现快速排序该算法，不允许直接调用插入排序函数，需要手动实现其逻辑过程。

```python
def quick_sort(L,start,end):
    if start >= end:
        return
    mid_data, left, right = L[start], start, end
    while left < right:
        while L[right] >= mid_data and left < right:
            right -= 1
        L[left] = L[right]
        while L[left] < mid_data and left < right:
            left += 1
        L[right] = L[left]
    L[left] = mid_data
    quick_sort(L, start, left-1)
    quick_sort(L, left+1, end)

L_str = input("请输入一组整数，用空格分隔: ")
L = list(map(int, L_str.split(' ')))
quick_sort(L,0,len(L)-1)
print("排序后:", L)
```

5.使用 Python 随机生成三组长度递增的随机数列，长度分别为100、1000、10000，对生成的随机数组使用不同的排序算法，例如选择排序和归并排序，至少三种排序方法。对这些数列的数据排序，并分析不同排序算法在不同长度数列下的运行效果，例如运行时间等指标。

```python
import random
import time

def random_array(n):#随机生成数组
    arr = []
    arr.append(random.randint(1, 10))
    for i in range(1,n):
        temp = random.randint(arr[i-1], n)
        arr.append(temp)
    return arr
    
def insertion_sort(L):#插入

    for i in range(1,len(L)):
        temp = L[i]
        j = i - 1
        while j >= 0 and L[j] > temp:
            L[j+1] = L[j]
            j -= 1
        L[j+1] = temp

def selction_sort(L):#选择
    for i in range(len(L)-1):
        min_index = i
        for j in range(i+1, len(L)):
            if L[j] < L[min_index]:
                min_index = j
        L[i], L[min_index] = L[min_index], L[i]
    return L


def quick_sort(L):#快排
    stack = [(0, len(L) - 1)]
    while stack:
        start, end = stack.pop()
        if start >= end:
            continue
        
        mid_data, left, right = L[start], start, end
        while left < right:
            while L[right] >= mid_data and left < right:
                right -= 1
            L[left] = L[right]
            
            while L[left] < mid_data and left < right:
                left += 1
            L[right] = L[left]
        
        L[left] = mid_data
        stack.append((start, left - 1))
        stack.append((left + 1, end))



def insertion_sort_time(L,n):
    sumtime = 0
    for i in range(1,100):
        random_array(n)
        start = time.time()
        insertion_sort(L)
        end = time.time()
        Time = end - start
        sumtime += Time
    return sumtime

def selction_sort_time(L,n):
    sumtime = 0
    for i in range(1,100):
        random_array(n)
        start = time.time()
        selction_sort(L)
        end = time.time()
        Time = end - start
        sumtime += Time
    return sumtime

def quick_sort_time(L,n):
    sumtime = 0
    for i in range(1,100):
        random_array(n)
        start = time.time()
        quick_sort(L)
        end = time.time()
        Time = end - start
        sumtime += Time
    return sumtime

n = 100
L = random_array(n)
arr1_insertion_sort_time = insertion_sort_time(L,n)
arr1_selction_sort_time = selction_sort_time(L,n)
arr1_quick_sort_time = quick_sort_time(L,n)

m = 1000
arr2_insertion_sort_time = insertion_sort_time(L,m)
arr2_selction_sort_time = selction_sort_time(L,m)
arr2_quick_sort_time = quick_sort_time(L,m)

k = 10000
arr3_insertion_sort_time = insertion_sort_time(L,k)
arr3_selction_sort_time = selction_sort_time(L,k)
arr3_quick_sort_time = quick_sort_time(L,k)

print("由于单次排序时间过短.所以求了循环100次的总时间")
print("长度100时:")
print("插入排序时间：", arr1_insertion_sort_time)
print("选择排序时间：", arr1_selction_sort_time)
print("快速排序时间：", arr1_quick_sort_time)
print("长度1000时:")
print("插入排序时间：", arr2_insertion_sort_time)
print("选择排序时间：", arr2_selction_sort_time)
print("快速排序时间：", arr2_quick_sort_time)
print("长度10000时:")
print("插入排序时间：", arr3_insertion_sort_time)
print("选择排序时间：", arr3_selction_sort_time)
print("快速排序时间：", arr3_quick_sort_time)

```
